# Matching Algorithm for Skill-Share Platform

## Project Overview
This project aims to develop a matching algorithm that connects users in a skill-sharing web application based on their possessed skills, desired skills, and their involvement in various collaborative projects. Each user provides details such as their personal skills (roles they can perform), desired skills (roles they want to learn or collaborate on), and their location. The goal of the algorithm is to efficiently match users with complementary skills to help them collaborate on projects.

We have already generated a test dataset containing 150 sample users with attributes such as name, age, project, personal skill, desired skills, and location. The matching algorithm will leverage this dataset to suggest optimal collaborators for each user.

## Collaborative Challenge
Each collaborator is tasked with contributing their own ideas and approaches to the development of the matching algorithm. Here are the key challenges:

1. Skill Matching Logic: Develop a logic that identifies users with complementary skills. A potential approach could involve filtering users who possess a skill another user is seeking and vice versa.
   
2. Project-Based Matching: Ensure that the matching takes into account the type of project each user is involved in, prioritizing matches within the same project type.
   
3. Location Proximity: Consider geographical location as a factor, giving higher priority to users who are in closer proximity to each other, if relevant to the project's needs.
